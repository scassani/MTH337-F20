To combine boolean arrays we can use
* `&` for `and`
* `~` for `not`
* `|` for `or`

**Ex.** build an array `c` which is `True` when the entry of `a` is even, and can be divided by 5

```python
d = a % 2 == 0 # array True when even
e = a % 5 == 0 # array Trus when divisible by 5
c= d & e
print(c)
```

```python
print(a[c]) # indexing using boolean arrays
```

```python
a[c==False]
```

```python
a[~c]
```
