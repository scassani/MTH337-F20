## File operations

`open()` opens a file (or it creates it), `write()` can be used to write in the file, `close()` is used to close the file

```python
f=open('grocery.txt','w') # open the file in writing mode
f.write('bread\n') # \n is a special character (newline)
f.write('bananas\n')
f.write('doritos\n')
f.write('takis\n')
f.write('eggs')
f.close()
```

```python
f=open('grocery.txt','a') # open the file in append mode (to add something to file)
```

```python
f.write('\n pizza')
f.close()
```

`read()` to read the file

```python
f=open('grocery.txt','r') # reading mode
```

```python
a=f.read()
print(a)
```

```python
b=f.read()
print(b)
```

```python
f.tell() # the reading cursor is stuck at the end of the file
```

To read the file again we need to reset the cursor at the beginning (`seek()`)

```python
f.seek(0) # back at the beginning
f.tell()
```

```python
c=f.read()
print(c)
```

```python
f.seek(0)
d=f.read(10) # will read the first 10 characters
print(d)
```

Other reading functions

```python
f.seek(0)
e=f.readlines() # reads the whole file and returns a list of lines
print(e)
```

```python
f.seek(0)
for line in f:
    print(line)
```

```python
f.seek(0)
for k in range(5):
    print(f.readline())
```

```python
f.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
imga=np.random.rand(5,4,3)
```

Q3: write the code that makes a copy of `imga` (call the copy `imgb` ), and changes to black all the pixels
where the red component of the image is greater than 0.5. Use boolean indexes.

```python
imgb=imga.copy()
```

```python
k=imgb[:,:,0]>0.5 # red above 0.5
```

```python
imgb[k] = 0
```

```python
plt.subplot(121)
plt.imshow(imga)
plt.subplot(122)
plt.imshow(imgb)
plt.show()
```

Q4: Write the code that makes a copy of `imga` (call the copy `imgc`), and changes to 1 the blue component
of all the pixels where the green component of the image is between 0.1 and 0.7. Use boolean indexes.

```python
imgc=imga.copy()
greenzone1=imgc[:,:,1]<.7 # boolean array
greenzone2=imgc[:,:,1]>.1 # boolean array
cond = greenzone1 & greenzone2
print(cond)
```

```python
cond.shape
```

```python
imgc[cond,2]=1
```

```python
plt.subplot(121)
plt.imshow(imga)
plt.subplot(122)
plt.imshow(imgc)
plt.show()
```

Q2: Write the code that creates a 3x4 array `a` with numbers from 0 to 11. Use boolean indexes to set
all of the entries of `a`, that are less than 2 or greater than 8, equal to 0.

```python
a=np.arange(12).reshape(3,4)
print(a)
```

```python
c1=a<2 # entries of a that are less than 2
c2=a>8
c=c1|c2

a[c]=0
print(a)
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import colors

x = np.zeros((200,200), dtype = int)

fig = plt.figure(figsize = (6,6))
ax = plt.subplot(111)
plt.title("Animation test")
cmap = colors.ListedColormap(['palegreen', 'green', 'black', 'orange'])
im = ax.imshow(x, cmap = cmap, vmin = -1, vmax = 2)

def animate(i):
    global x

    x=np.random.random(x.shape)
    im.set_data(x)
    return im


anim = FuncAnimation(fig=fig, func=animate,  interval=100, blit=True, repeat=False)
plt.show()
```
